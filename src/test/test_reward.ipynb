```python

# Reward function
def compute_reward_value(reward: int | float, reward_type: str) -> int | float:
    if reward_type == "cr":
        return 1 - int(reward)  # invert CR
    if reward_type == "is_acknowledgement":
        return int(reward)
    if reward_type == "is_grammatical":
        return int(reward)
    if reward_type.startswith("align"):
        return int(reward > 0)
    if reward_type.startswith("sent"):
        return reward
    raise RuntimeError("Unrecognized reward type: " + reward_type)

# Mapping of columns to reward types
column_reward_mapping = {
    "is_cr": "cr",
    "is_acknowledgement": "is_acknowledgement",
    "is_grammatical": "is_grammatical",
    "align_semantic": "align_semantic",
    "align_lexical_unigram": "align_lexical_unigram",
    "align_lexical_bigram": "align_lexical_bigram",
    "align_syntactic": "align_syntactic",
    "sent_approval": "sent_approval",
    "sent_caring": "sent_caring",
    "sent_curiosity": "sent_curiosity",
    "sent_engagement": "sent_engagement",
    "sent_negativity": "sent_negativity",
    "sent_supportiveness": "sent_supportiveness",
    "sent_warmth": "sent_warmth"
}

# Apply the function to all relevant columns
for col, reward_type in column_reward_mapping.items():
    if col in data.columns:
        data[f"{col}_reward"] = data[col].apply(lambda x: compute_reward_value(x, reward_type))

# Inspect results
print(data.head())

# Optionally save to CSV
data.to_csv("reward_test_results.csv", index=False)
```
